```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

#Load data from CSV

file_path = r'q3.csv'

df = pd.read_csv(file_path)

print(df.head())
```

```python
X = np.array(df.iloc[:,1]).reshape(-1,1)
Y = np.array(df.iloc[:,2]).reshape(-1,1)

reg = LinearRegression().fit(X,Y)
```

```python
import matplotlib.pyplot as plt
# Plot outputs
y_predict = reg.predict(X)
plt.scatter(X, Y, color="black")
plt.plot(X,y_predict)

plt.show()
```

```python
def loss(y,predicted_y,beta):
    total = np.sum(((y-predicted_y)**2)/(((y-predicted_y)**2)+beta**2))
    N = y.shape[0]
    return total/N
```

```python
Model01_Loss = loss(Y,-4*X+12,1)
Model02_Loss = loss(Y,-3.55*X+3.91,1)

print("Model 01 Loss:",Model01_Loss)
print("Model 02 Loss:",Model02_Loss)
```

```python
# Plot outputs
y_predict_model01 = -4*X+12,1
plt.scatter(X, Y, color="black")
plt.plot(X,y_predict_model01[0])

plt.show()
```

```python
# Plot outputs
y_predict_model02 = -3.55*X+3.91
plt.scatter(X, Y, color="black")
plt.plot(X,y_predict_model02)

plt.show()
```

```python
#Analysing the impact of beta

for b in range(0,100,10):
    Model01_Loss = loss(Y,-4*X+12,b)
    Model02_Loss = loss(Y,-3.55*X+3.91,b)

    print("Beta:",str(b))
    print("Model 01 Loss:",Model01_Loss)
    print("Model 02 Loss:",Model02_Loss)
```

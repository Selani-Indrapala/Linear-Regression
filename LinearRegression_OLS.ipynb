This is an implementation of linear regression using the **Ordinary Least Mean Square Approach** to find coefficients.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data(
                                                                path="boston_housing.npz", test_split=0.2, seed=113
                                                            )
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
plt.scatter(x_train[:,12], y_train, color='red', label='Data')
```

```python
plt.scatter(x_train[:,5], y_train, color='red', label='Data')
```

```python
#Calculating the coefficients
def Coefficients(X,Y):
  n = X.shape[0]
  x_mean = np.mean(X)
  y_mean = np.mean(Y)
  num = 0
  den = 0
  for i in range(n):
      num += (X[i] - x_mean) * (Y[i] - y_mean)
      den += (X[i] - x_mean) ** 2

  w1 = num / den
  w0 = y_mean - (w1 * x_mean)
  return w0,w1
```

```python
#Regression Equation for 5th feature
w0,w1 = Coefficients(x_train[:,5],y_train)
print("y5 =",str(w0),"+",str(w1)+"x")
```

```python
#Plots

plt.scatter(x_train[:,5], y_train, color='red', label='Data')
plt.plot(x_train[:,5],w0+w1*x_train[:,5],color="blue",label="Regression Line")

plt.legend()
plt.show()
```

```python
#Regression Equation for 12th feature
w0,w1 = Coefficients(x_train[:,12],y_train)
print("y12 =",str(w0),"+",str(w1)+"x")
```

```python
#Plots

plt.scatter(x_train[:,12], y_train, color='red', label='Data')
plt.plot(x_train[:,12],w0+w1*x_train[:,12],color="blue",label="Regression Line")

plt.legend()
plt.show()
```
